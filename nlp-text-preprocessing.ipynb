# **Lowercasing**

```python
df['review']=df['review'].str.lower()
```

# Step 2- Remove HTML tag

```python
import re
```

```python
def remove_tag (text):
    pattern=re.compile('<.*?>')
    return pattern.sub(r'',text)
```

```python
df['review']=df['review'].apply(remove_tag)
```

```python
df.head()
```

# Step 3- Remove URl 

```python
def remove_url(text):
    pattern=re.compile(r'https ? ://\s+|www\.\s+')
    return pattern.sub(r'',text)
```

```python
df['review']=df['review'].apply(remove_url)
```

```python
df.head()
```

# Step 4 - Remove Punctuation

```python
import string
exclude=string.punctuation
```

```python
def remove_punc(text):
    for char in exclude:
        text=text.replace(char,'')
        return text
```

```python
df['review']=df['review'].apply(remove_punc)
```

```python
def chat_words(text):
    new_text=[]
    for w in text.split():
        if w.upper()in chatword:
            new_text.append(chatword[w.upper()])
        else:
            new_text.append(w)
    return " ".join(new_text)
```

```python
chat_words(test2)
```

```python
df['review']=df['review'].apply(chat_words)
```

# Step 6- Remove Stop Words

```python
from nltk.corpus import stopwords
```

```python
stopwords.words('english')
```

```python
def remove_stopW(text):
    new_text=[]
    for word in text.split():
        if word in stopwords.words('english'):
            new_text.append('')
        else:
            new_text.append(word)
    x=new_text[:]
    new_text.clear()
    return " ".join(x)
```

```python
df['review']=df['review'].apply(remove_stopW)
```

```python
test3='a market is bad'
remove_stopW(test3)
```
